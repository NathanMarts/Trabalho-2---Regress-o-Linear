```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

%matplotlib inline
```

```python
df = pd.read_csv('AirQualityUCI.csv', sep = ';')
df = df.drop(columns=['Date', "Time"])
df.head()
```

```python
df_norm = (df - df.min()) / (df.max() - df.min())
df_norm.head()
```

```python
n_features = len(df_norm.columns)-1
n_features
```

```python
w = np.random.rand(1,n_features+1) ## Gerando valores aleatórios entre 0 e 1
w
```

```python
X = np.array(df_norm.drop('AH', axis=1))

y = df_norm.iloc[:, 12:13].values
```

```python
def insert_ones(X):
    ones = np.ones([X.shape[0],1])
    return np.concatenate((ones,X) , axis=1)
```

```python
def custo(w,X,y):
    m = len(X)
    
    erro = (X@w.T - y)
    
    custo = np.sum(np.power(erro,2))
    
    return custo/m
```

```python
def gradient_descent(w, X, y, iteracoes):
    cost = np.zeros(iteracoes)
    for i in range(iteracoes):
        w = w - (1/len(X)) * np.sum((X@w.T - y)*X, axis=0)
        
        cost[i] = custo(w, X, y)
    
    return w, cost
```

```python
X = insert_ones(X)
```

```python
w, cost = gradient_descent(w, X, y, 95)
```

```python
custo(w, X, y) #custo final
```

```python
#plotando o custo
fig, ax = plt.subplots()  
ax.plot(np.arange(95), cost, 'r')  
ax.set_xlabel('Iterações')  
ax.set_ylabel('Custo')  
ax.set_title('Erro vs. iterações')
```
