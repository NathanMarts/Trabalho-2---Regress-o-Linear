```python
X = np.array(df_norm.drop('AH', axis=1))

y = df_norm.iloc[:, 12:13].values
```

```python
def insert_ones(X):
    ones = np.ones([X.shape[0],1])
    return np.concatenate((ones,X) , axis=1)
```

```python
def custo(w,X,y):
    m = len(X)
    
    erro = (X@w - y)
    
    custo = np.sum(np.power(erro,2))
    
    return custo/m
```

```python
def gradient_descent(w, X, y, epoch):
    cost = np.zeros(epoch)
    
    for i in range(epoch):
        
        w = w - (1/len(X)) * np.sum((X@w - y)*X, axis=0)
        
        cost[i] = custo(w, X, y)
    
    return w, cost
```
